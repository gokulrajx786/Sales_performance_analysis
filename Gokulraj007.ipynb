```python
import numpy as np
import pandas as pd

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import pandas as pd
df = pd.read_csv('/content/Real Estate Sales.csv', low_memory=False)
df.head()

df.isna().sum()

df_copy = df.drop(columns=['OPM remarks','Non Use Code','Assessor Remarks','Location','Residential Type'])
df_copy.head()

df_copy.isna().sum()

df_copy = df_copy.fillna(df.mode().iloc[0])
df_copy.isna().sum()

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
x = df_copy['Property Type'].value_counts().head(6)
plt.pie(x=x, labels=x.index, autopct='%1.1f%%',data=df_copy);

plt.subplot(1,2,2)
x = df_copy['Town'].value_counts().head(20)
plt.pie(x=x, labels=x.index, autopct='%1.1f%%', data=df_copy);

plt.figure(figsize=(20,7))
x = df_copy['Town'].value_counts().head(30)
plt.plot(x,'-ok')

plt.xticks(rotation=90);

plt.figure(figsize=(15,7))
plt.hist(df_copy['Property Type'],histtype='stepfilled', bins=11,density=True, color='green',orientation='horizontal')
plt.title('Property Type Histogram Plot')
plt.show()

plt.figure(figsize=(10,7))
plt.plot(df['Assessed Value'],'r-',alpha=0.5,label='AV')
plt.plot(df['Sale Amount'],'g-',alpha=0.5, label='SA')
plt.title('Assessed Value vs Sale Amount')
plt.legend()
plt.show()

plt.figure(figsize=(14,7))
x = df_copy['Sales Ratio']
plt.plot(x)
plt.show()

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Real Estate Sales.csv')

plt.figure(figsize=(10, 6))
plt.scatter(data['Assessed Value'], data['Sales Ratio'], color='blue', alpha=0.6)

plt.title("Real Estate Sales: Sales Ratio vs. Assessed Value")
plt.xlabel("Assessed Value")
plt.ylabel("Sales Ratio")
plt.grid(True)

plt.show()

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Real Estate Sales.csv')

avg_price = data.groupby('Town')['Sale Amount'].mean()

plt.figure(figsize=(12, 8))
plt.bar(avg_price.index, avg_price.values, color='blue')

plt.title("Average Real Estate Sale Price")
plt.xlabel("Town")
plt.ylabel("Sales Amount")
plt.show()

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Real Estate Sales.csv')

plt.figure(figsize=(10, 6))
plt.hist(data['Assessed Value'], bins=20, color='blue', alpha=0.7)

plt.title("Distribution of Real Estate Sale Prices")
plt.xlabel("Assessed Value")
plt.ylabel("Sales Ratio")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()






```
